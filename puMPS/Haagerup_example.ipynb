## I. Ground state optimization (N=6) 

```julia
N = 6 #number of anyons 
D = 12 #bond dimenson (must be multiple of 3) #Z3 symmetry is explicit
```

```julia
M_full = puMPS.puMPState(randn(Complex{Float64},D,6,D),N); #random MPS, d=6 is the Haagerup local dimension
```

```julia
leg1 = SymLeg(3,[0,1,2],[div(D,3),div(D,3),div(D,3)])
leg2 = SymLeg(3,[0,1,2],[2,2,2])
leg3 = conj(leg1);
legs = SymLeg[leg1,leg2,leg3];
M = sym_puMPS_from_dense(M_full, legs);
normalize!(M); #transforming into a symmetric MPS
```

```julia
H_MPO_full,H_MPO = Haagerup_local_MPO_sym(); #Haagerup Hamiltonian MPO
```

```julia
maxiters = 2000 #number of iteration
@time M,normgrad = minimize_energy_local_sym!(M, H_MPO, maxiters, tol = 1E-4);
```

## II. Excited state with fixed momentum(N=6,p=0)

```julia
function runexcitation(M,ks)
    d = 6
    M_full = full(M)
    
    legs = M.A.legs
    M_full, lambda, lambda_i = canonicalize_left_sym!(M_full,legs)

    H_full_PBC = Haagerup_PBC_MPO_sym_split();

    num_states = fill(8,length(ks)) #Compute 8 states per momentum

    Gs, Heffs = puMPS.tangent_space_metric_and_MPO(M_full, H_full_PBC, ks, eye(Complex{Float64},puMPS.bond_dim(M_full)));
    
    puMPS.tspace_ops_to_center_gauge!(Gs, lambda_i)
    puMPS.tspace_ops_to_center_gauge!(Heffs, lambda_i)
    
    ens, exs = excitations_hybrid(M_full,ks,Gs,Heffs,lambda_i,num_states);

    return M_full,ens,exs
end
```

```julia
ks = [0,2] #momentum to compute
M_full, ens, exs = runexcitation(M,ks); #ens are energies, exs are the exitation tensors
```

```julia
save("Haagerup_puMPS_exs_N$(N)D$(D).jld","A",M_full.A,"ks",ks,"ens",ens,"exs",exs);  #We can generate all excitation data in this way
```

## III. Compute ground state overlap II->I (N=6,N=6->N=12) using existing data

```julia
function compute_ovs(N1::Int,N2::Int,D1::Int,Ds2::Vector{Int})
    ## ground state overlap
    A1 = load("Haagerup_sympuMPS_N$(N1)D$(D1).jld","A");
    M1 = puMPS.puMPState(A1,N1)
    ovs = zeros(length(Ds2))
    for i in 1:length(Ds2)
        D2 = Ds2[i]
        A2 = load("Haagerup_sympuMPS_N$(N2)D$(D2).jld","A");
        M2 = puMPS.puMPState(A2,N2);
        @time ovi = abs(overlap_3state(M1,M1,M2))
        ovs[i] = ovi
    end
    return ovs
end
```

```julia
compute_ovs(6,12,24,[51])
```

## IV. Compute ground state to excitation overlap II->epsilon (N=6,N=6->N=12) using existing data

```julia
function compute_IIep(N1::Int,N2::Int,N3::Int,D1::Int,D2::Int,Ds3::Vector{Int})
    A1 = load("Haagerup_sympuMPS_N$(N1)D$(D1).jld","A");
    M1 = puMPS.puMPState(A1,N1)
    A2 = load("Haagerup_sympuMPS_N$(N2)D$(D2).jld","A");
    M2 = puMPS.puMPState(A2,N2);
    
    ovs = zeros(length(Ds3))
    for i in 1:length(Ds3)
        D3 = Ds3[i]
        A3 = load("Haagerup_puMPS_exs_N$(N3)D$(D3).jld","A");
        B3 = load("Haagerup_puMPS_exs_N$(N3)D$(D3).jld","exs")[2][2]; #The indices are [momentum index (p=0)][Excited state index]
        J3Bp = zero(A3)
        J3Bp[sym_inds_to_dense(D3)] = B3
        ex3 = puMPS.puMPSTvec(puMPS.puMPState(A3,N3),J3Bp,0.0);
        @time ovi = abs(overlap_3state(M1,M2,ex3))
        ovs[i] = ovi
    end
    return ovs
end
```

```julia
compute_IIep(6,6,12,12,12,[51])
```

## V. Compute overlap JJ*->I (N=7,N=8 to N=15), using existing data

```julia
function compute_JJI(N1::Int,N2::Int,N3::Int,D1::Int,D2::Int,Ds3::Vector{Int})
    A1 = load("Haagerup_puMPS_exs_N$(N1)D$(D1).jld","A");
    B1 = load("Haagerup_puMPS_exs_N$(N1)D$(D1).jld","exs")[4][1]; #[Momentum s=1 label][ground state label]
    J1Bp = zero(A1)
    J1Bp[sym_inds_to_dense(D1)] = B1
    ex1 = puMPS.puMPSTvec(puMPS.puMPState(A1,N1),J1Bp,1.0);

    A2 = load("Haagerup_puMPS_exs_N$(N2)D$(D2).jld","A");
    B2 = load("Haagerup_puMPS_exs_N$(N2)D$(D2).jld","exs")[4][1];#[Momentum s=1 label][ground state label]
    J2Bp = zero(A2)
    J2Bp[sym_inds_to_dense(D2)] = B2
    ex2 = puMPS.puMPSTvec(puMPS.puMPState(A2,N2),J2Bp,1.0);
    
    ovs = zeros(length(Ds3))
    for i in 1:length(Ds3)
        D3 = Ds3[i]
        A3 = load("Haagerup_sympuMPS_N$(N3)D$(D3).jld","A");
        M3 = puMPS.puMPState(A3,N3);
        @time ovi = abs(overlap_3state(ex1,ex2,M3))
        ovs[i] = ovi
    end
    return ovs
end
```
